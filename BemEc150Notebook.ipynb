```python
import pandas as pd
from scipy import signal
import matplotlib.pyplot as plt
from pylab import figure
from scipy.fftpack import fft, ifft

%matplotlib inline
df = pd.read_csv('exoTrain.csv')
example_exoplanet_1 = df.iloc[0].as_matrix()
example_exoplanet_2 = df.iloc[1].as_matrix()
example_notexoplanet_1 = df.iloc[100].as_matrix()
example_notexoplanet_2 = df.iloc[101].as_matrix()
```

```python
exoplanet_range_1 = range(1, len(example_exoplanet_1))
exoplanet_range_2 = range(1, len(example_exoplanet_2))
notexoplanet_range_1 = range(1, len(example_notexoplanet_1))
notexoplanet_range_2 = range(1, len(example_notexoplanet_2))


# First, design the Buterworth filter
N  = 4    # Filter order
Wn = 0.01 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')

# Second, apply the filter
example_exoplanet_1_filtered = signal.filtfilt(B,A, example_exoplanet_1[1:])
example_exoplanet_2_filtered = signal.filtfilt(B,A, example_exoplanet_2[1:])

# First, design the Buterworth filter
N  = 4    # Filter order
Wn = 0.01 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')

# Second, apply the filter
example_notexoplanet_1_filtered = signal.filtfilt(B,A, example_notexoplanet_1[1:])
example_notexoplanet_2_filtered = signal.filtfilt(B,A, example_notexoplanet_2[1:])

figure(1)
plt.title('Time Series Light Intensity for Star with Exoplanet')
plt.plot(exoplanet_range_1, example_exoplanet_1[1:], 'b', label = 'raw signal')
plt.plot(exoplanet_range_1, example_exoplanet_1_filtered, 'r', label = 'low pass filtered')
plt.legend(loc='best')
# plt.show()
plt.savefig('exoplanet_1.jpg')
figure(2)
plt.title('Time Series Light Intensity for Star without Exoplanet')
plt.plot(notexoplanet_range_1, example_notexoplanet_1[1:], 'b', label = 'raw signal')
plt.plot(notexoplanet_range_1, example_notexoplanet_1_filtered, 'r', label = 'low pass filtered')
plt.legend(loc='best')
# plt.show()
plt.savefig('no_exoplanet_1.jpg')

figure(3)
plt.title('Time Series Light Intensity for Star with Exoplanet')
plt.plot(exoplanet_range_2, example_exoplanet_2[1:], 'b', label = 'raw signal')
plt.plot(exoplanet_range_2, example_exoplanet_2_filtered, 'r', label = 'low pass filtered')
plt.legend(loc='best')
# plt.show()
plt.savefig('exoplanet_2.jpg')
figure(4)
plt.title('Time Series Light Intensity for Star without Exoplanet')
plt.plot(notexoplanet_range_2, example_notexoplanet_2[1:], 'b', label = 'raw signal')
plt.plot(notexoplanet_range_2, example_notexoplanet_2_filtered, 'r', label = 'low pass filtered')
plt.legend(loc='best')
# plt.show()
plt.savefig('no_exoplanet_2.jpg')
```
